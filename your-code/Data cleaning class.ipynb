```python
#Get rid of all the columns with nan he line of accepted nans is generally around 10%
ds_null = ds.isnull().sum()
ds_null [ds_null > 0] # Introducing filter!!! Adding function,
#in pandas you add filters and conditions into brackets
```

```python
#Check for zero values in the dataframe, then in particular columns
zeros_values = ds[ds == 0.0].count()
zeros_values [['cylinders', 'displ']]
```

```python
#Make a list to drop
drop_cols = list(ds_null[ds_null > 10000].index)     ## The threshold is a bit lax.... #we dont care about the value itseflf but about the index (cad name of the column)
ds_dropped = ds.drop(drop_cols, axis=1)
#check in place
ds_dropped.head()
```

```python
# Cylinders and displ have similar number of Null values? Maybe they have the same null entries? 
# We could drop them or keep them
null_displ = ds[ds['displ'].isnull()]

null_displ = null_displ[['year', 'make', 'model', 'trany', 'drive','fuelType','cylinders', 'displ']]
null_displ
```

```python
# In case we decide they are important even with little data,
# we might want to fill the missing values with a value we choose. In this case we fill zero. because all the zeros already n the dataframe are 0.0 so its easier tospot them later
ds[['displ', 'cylinders']] = ds[['displ', 'cylinders']].fillna(0)
ds[['year', 'make', 'model', 'trany', 'drive','fuelType','cylinders', 'displ']][29670:30000]
```

```python
# Is there any entry that has one of the two values different from 0?
#are then two (120 and 123) overlapping? in this cas yes
test = ds[(ds['cylinders']==0) & (ds['displ']!=0)]
test[['year', 'make', 'model', 'trany', 'drive','fuelType','cylinders', 'displ']]
```

```python
# We could replace the value with 4?  replace values in scepific cells for specific columns
#ds.loc[conditions to select all the rows where it happens , columnname]
ds.loc[(ds['cylinders']==0) & (ds['displ']!=0), 'cylinders'] = 4

test = ds[(ds['cylinders']==4) & (ds['displ']!=0)]
test[['year', 'make', 'model', 'trany', 'drive','fuelType','cylinders', 'displ']]
```
